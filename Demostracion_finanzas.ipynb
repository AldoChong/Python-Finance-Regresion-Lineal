El siguiente notebook es una demostracion de las capacidades de los metodos definidos en la clase finanzas.py.Tomaremos información de una empresa del IPC de México y calcularemos la Beta y sus razones financieras. Usaremos informacion del 2022 de Becle S.A.B. de C.V.

Empezamos calculando las razones financieras de la empresa. Son medidas que indican la salud financiera de una empresa. Mientras mejores razones tengan, mas atractivo sera invertir en acciones de dicha empresa. 

```python
#Empezamos importando la clase finanzas.py y llamamos uno de sus metodos
from finanzas import Razones
import pandas as pd

razones = pd.read_csv('razones.csv')
```

```python
#Inicializamos el objeto razones y llamamos a uno de sus metodos 

raz = Razones()
raz.summary('CUERVO')
```

Podemos hacer un análisis rápido de la salud financiera de la empresa. Vemos que tiene capacidad de pago de sus obligaciones a corto plazo por que su razon corriente es de 3.13. Tiene un Margen de utilidad neta de 0.47, es decir, sus utilidades son casi la mitad de sus ventas totales. Por otra parte el ROA es de 0.06 y el ROE es de 0.10, que son medidas relativamente buenas de rendimiento. Es decir, la administracion pudo generar un rendimiento de 0.06 sobre activos y 0.10 sobre capital. Se tiene que se obtuvo 1.6 pesos de utilidad por cada accion y se dio 0.13 pesos por accion de dividendos. En general, parece que la empresa tiene salud financiera y puede ser una buena opcion de inversion. 

Ahora pasamos a calcular la Beta de la accion. Recordemos que la Beta indica el grado de volatidad del rendimiento de la accion vs el rendimiento del mercado. Para calcularla, utilizamos el metodo de regresion lineal. Sea $Rend\_Accion$ el rendimiento de la accion y sea $Rend\_Market$ el rendimiento del mercado. Entonces la recta de regresion lineal es aquella que minimiza el error cuadratico medio entre los puntos. De tal suerte que $\hat{Rend\_Accion} = \beta_{0} + \beta_{1} Rend\_Market$ es la estimacion del rendimiento de la accion. La beta buscada es $\beta_{1}$. 


```python
#Importamos el modulo de beta y utilizamos sus metodos
from finanzas import Beta

beta = Beta('CUERVO.MX.csv', '^MXX.csv')
beta.summary()
```

Vemos que la Beta es de 0.54. Sea $\beta$ la Beta de la accion. Recordemos que si $\beta = 1$ entonces la accion fue tan volatil como el mercado. Si $\beta<1$ entonces la Beta 
fue menos volatil que el mercado y si $\beta>1$ entonces la Beta fue mas volatil que el mercado. De tal suerte que estamos frente a una accion que fue menos volatil que el mercado en 2022. En resumen, 

1. La empresa tiene buena salud financiera. 
2. La empresa tiene baja volatilidad. 

De tal suerte que si tenemos un perfil de inversion conservador, enfocado a la construccion de capital, con dividendos como fuentes de ingreso y ganancias de capital en el mediano o largo plazo (1 año o mas), entonces podemos considerar como buena opcion agregar acciones de esta empresa a nuestra cartera de inversion.
